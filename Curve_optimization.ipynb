```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
np.random.seed(42)  # For reproducibility
```

```python
Samples =  300
Layers = [1,10,10,10,1]
LR = 0.5
epochs = 10000
```

```python
x_samples = np.random.uniform(low=0, high=10*np.pi, size=(Samples,1)) #row x column
y_samples = np.sin(x_samples) + np.random.normal(loc=0.0, scale=0.5, size=(Samples,1)) 
```

```python
plt.scatter(x_samples, y_samples, color='blue', label='Data Samples')
plt.title('Sample Data Points')
```

```python
pd.DataFrame({'x': x_samples.flatten(), 'y': y_samples.flatten()}).head()
```

```python
sigmoid = lambda x: 1 / (1 + np.exp(-x))
identity = lambda x: x

#defined the activation functions
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
# Build a model that converges easily for the given data
model = keras.Sequential([
    keras.layers.Dense(50, input_shape=(1,), activation='relu'),
    keras.layers.Dense(50, activation='relu'),
    keras.layers.Dense(50, activation='relu'),
    keras.layers.Dense(50, activation='relu'),
    keras.layers.Dense(50, activation='relu'),
    keras.layers.Dense(1)
])

model.compile(optimizer=keras.optimizers.Adam(learning_rate=0.05), loss='mse', metrics=['mae'])
model.fit(x_samples, y_samples, epochs=500)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Normalize x between 0 and 1 for better convergence
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x_samples)

model1 = keras.Sequential([
    keras.layers.Dense(64, input_shape=(1,), activation='tanh'),
    keras.layers.Dense(64, activation='tanh'),
    keras.layers.Dense(64, activation='tanh'),
    keras.layers.Dense(1)
])

model1.compile(optimizer=keras.optimizers.Adam(learning_rate=0.001),
              loss='mse', metrics=['mae'])

model1.fit(x_scaled, y_samples, epochs=1000, batch_size=32)
```

```python
# Predict using the trained model
y_pred = model1.predict(x_samples)

# Plot original data and model predictions
plt.figure(figsize=(10, 6))
plt.scatter(x_samples, y_samples, color='blue', label='Original Data')
plt.scatter(x_samples, y_pred, color='red', alpha=0.6, label='Model Prediction')
plt.title('Model Fit vs Original Data')
plt.xlabel('x_samples')
plt.ylabel('y_samples')
plt.legend()
plt.show()
```

```python
from tensorflow.keras import layers
class Sine(layers.Layer):
    def call(self, inputs):
        return tf.sin(inputs)

# SIREN weight initializer
def sine_init(shape, dtype=None):
    return tf.random.uniform(shape, minval=-1.0, maxval=1.0, dtype=dtype)

model2 = keras.Sequential([
    layers.Dense(64, input_shape=(1,), kernel_initializer=sine_init),
    Sine(),
    layers.Dense(64, kernel_initializer=sine_init),
    Sine(),
    layers.Dense(64, kernel_initializer=sine_init),
    Sine(),
    layers.Dense(1)
])

model2.compile(optimizer=keras.optimizers.Adam(learning_rate=0.001), 
              loss='mse', metrics=['mae'])

model2.fit(x_samples, y_samples, epochs=2000, batch_size=32)
```

```python
# Predict using the trained model
y_pred = model2.predict(x_samples)

# Plot original data and model predictions
plt.figure(figsize=(10, 6))
plt.scatter(x_samples, y_samples, color='blue', label='Original Data')
plt.scatter(x_samples, y_pred, color='red', alpha=0.6, label='Model Prediction')
plt.title('Model Fit vs Original Data')
plt.xlabel('x_samples')
plt.ylabel('y_samples')
plt.legend()
plt.show()
```
